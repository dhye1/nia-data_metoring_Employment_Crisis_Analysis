- 지역(시군구)만 추출

```python
job_df['(지역별)시군구'].value_counts()
```

```python
print('(지역별)시군구 수:', len(job_df['(지역별)시군구'].value_counts()))
```

```python
job_si_df = job_df['(지역별)시군구'].str.split()
job_si_df
```

```python
## (지역별)시군구 종류 확인
job_si_df.value_counts().index.to_list()
```

```python
job_si_df
```

```python
## 지역(시군구)만 추출
job_df['시군구'] = [si[0] if len(si) == 1 else si[1] for si in job_si_df]
```

```python
job_df.head(3)
```

```python
job_df['시군구'].value_counts().index.to_list()
```

```python
job_df.isnull().sum()
```

- 직종 수정

```python
job_df['직종_소분류'].value_counts() ## 137개의 직종_소분류 존재
```

```python
## 앞에 2018_직종 문구 없애기
job_df['직종_소분류'] = job_df['직종_소분류'].str.replace(r'2018직종_', '', regex=False)
```

```python
job_df
```

```python
## 직종 특수문자 제거하기 ? => / 로 변경
job_df['직종_소분류'] = job_df['직종_소분류'].str.replace(r'?', '/', regex=False)
```

```python
job_df
```

```python
## 수정한 직종_소분류 확인
list(job_df['직종_소분류'].value_counts().index)
```

```python
## 직종 \u3000 특수문자 제거하기 
job_df['직종_소분류'] = job_df['직종_소분류'].str.replace(r'\u3000', '')
```

```python
job_df
```

```python
list(job_df['직종_소분류'].value_counts().index)
```

- 데이터프레임 저장

```python
save_path = '/content/drive/MyDrive/데멘토/dataset/'
job_df.to_csv(save_path + '구인구직취업현황_전국_전처리완료.csv', index=False)
```
